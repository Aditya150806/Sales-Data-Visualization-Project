```python
# ============================================
# SALES DATA CLEANING & VISUALIZATION PROJECT
# ============================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------------------------
# STEP 1: LOAD DATASET
# --------------------------------------------

df = pd.read_csv("sales_data.csv")

print("FIRST 5 ROWS")
print(df.head())

# --------------------------------------------
# STEP 2: BASIC INFORMATION
# --------------------------------------------

print("\nDATASET INFO")
print(df.info())

print("\nMISSING VALUES")
print(df.isnull().sum())

print("\nSTATISTICAL SUMMARY")
print(df.describe())

# --------------------------------------------
# STEP 3: DATA CLEANING
# --------------------------------------------

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Fill missing values if any
df.fillna(df.mean(numeric_only=True), inplace=True)

print("\nMISSING VALUES AFTER CLEANING")
print(df.isnull().sum())

# --------------------------------------------
# STEP 4: DATA VISUALIZATION
# --------------------------------------------

sns.set(style="whitegrid")

# 1. Direction Count
plt.figure(figsize=(6,4))
sns.countplot(x='Direction', data=df)
plt.title("Market Direction Count")
plt.show()

# 2. Volume Distribution
plt.figure(figsize=(8,5))
plt.hist(df['Volume'], bins=20)
plt.title("Volume Distribution")
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.show()

# 3. Today's Return Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Today'], bins=30, kde=True)
plt.title("Today's Return Distribution")
plt.show()

# 4. Correlation Heatmap
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# 5. Boxplot for Volume
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Volume'])
plt.title("Boxplot for Volume")
plt.show()

# --------------------------------------------
# STEP 5: SAVE CLEANED DATA
# --------------------------------------------

df.to_csv("cleaned_sales_data.csv", index=False)

print("\nCLEANED DATASET SAVED SUCCESSFULLY!")

# --------------------------------------------
# STEP 6: INSIGHTS
# --------------------------------------------

print("\nPROJECT INSIGHTS:")
print("1. Market direction varies across days.")
print("2. Volume data contains some outliers.")
print("3. Most return values are concentrated near zero.")
print("4. Heatmap shows correlations between variables.")
print("5. Data cleaning improved dataset consistency.")
```
